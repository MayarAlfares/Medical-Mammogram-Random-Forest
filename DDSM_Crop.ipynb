**For Positive (Benign and Malignant) Mammograms ROI:**

```python
import cv2
import matplotlib.pyplot as plt 
import os
```

```python
def cropRect( mask ):
    contours, hierarchy = cv2.findContours(mask, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    contours = sorted(contours, key=lambda x: cv2.contourArea(x) )[-1]
    x,y,w,h = cv2.boundingRect(contours)
    return x,y,w,h

def crop( image, x,y,w,h ):return image[y:y+h, x:x+w]
```

```python
#test for 1 image
            
im_path = 'C:/Users/victory/Downloads/MINI-DDSM-Complete-JPEG/Cancer/0001/C_0001_1.RIGHT_MLO.jpg'
mask_path = 'C:/Users/victory/Downloads/MINI-DDSM-Complete-JPEG/Cancer/0001/C_0001_1.RIGHT_MLO_Mask.jpg'
save_path = 'D:/DDSM Dataset/Normal/'
#print(mask_path[:-9] + ".jpg")

im = cv2.imread(im_path)
mask = cv2.imread(mask_path)
mask = cv2.cvtColor(mask, cv2.COLOR_RGB2GRAY)
x,y,w,h = cropRect(mask)
res = crop( im, x,y,w,h )

filename = save_path + "_ROI_" + "1" + ".jpg" #index here

#cv2.imwrite(filename, res)
#imgplot = plt.imshow(res)
```
